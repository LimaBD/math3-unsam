```python
from regression import Point, LinearRegression

points = [Point(21, 20),
          Point(35, 33),
          Point(25, 25),
          Point(38, 40),
          Point(30, 35),
          Point(28, 32),
          Point(36, 37),
          Point(23, 25),
          Point(26, 30),
          Point(32, 30)]

linear = LinearRegression(points)

# NOTE: values could be rounded
print(('LINEAR REGRESSION RESULTS\n'
       f'Slope:\t\t\t{linear.get_slope()}\n'
       f'Y interception:\t\t{linear.get_y_interception()}\n'
       f'Square error:\t\t{linear.get_square_error()}'))

linear.graph()
```
